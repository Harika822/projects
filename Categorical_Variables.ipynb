Working With Categorical Values

```
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```
fruit = pd.read_csv('fruit_data_with_colors.txt',delimiter='\t')
fruit.head(5)
```

Label Encoder

```
nums = LabelEncoder()
fruit['fruit_name']=nums.fit_transform(fruit['fruit_name'].astype('str'))
fruit.head()
```

Dummy Coding


```
fruit_types= fruit[['fruit_name','mass','height','width']]
specific_fruit =pd.get_dummies(fruit_types)
fruit.head(5)
```

```
specific_fruit.head(5)
```

```
specific_fruit.shape
```

```
my_fruit = fruit[['fruit_label','fruit_name','fruit_subtype','mass','height','width']]
my_fruit.head(5)
```

```
dummy_fruit =pd.get_dummies(my_fruit)
dummy_fruit.head(5)
```

One Hot Encoder(OHE)

```
fruit_ohe = fruit[['fruit_label','mass']]
fruit_ohe.shape
```

```
enc=OneHotEncoder()
fruit_onehot =enc.fit_transform(fruit_ohe).toarray()
fruit_onehot.shape
```

```

label_onehot = fruit[['fruit_name']]
nums = LabelEncoder()
label_onehot['fruit_name']=nums.fit_transform(label_onehot['fruit_name'].astype('str'))

print(label_onehot.shape)
label_onehot.head(5)
```

```

ohe=OneHotEncoder()
train_ohe = ohe.fit_transform(label_onehot).toarray()
print(label_onehot.shape)
train_ohe
```
